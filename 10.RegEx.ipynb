A regular expression is a sequence of characters that define a search pattern. Usually such patterns are used by string-searching algorithms for "find" or "find and replace" operations on strings, or for input validation.

```
import re
```

```
sentence = 'My name is vimukthini. My mobile number is 0716275565 and my home number is 0312235449. We are living in year 2021 and today is the 25th of November'
```

```
#defining the search pattern
#\d denote a number
phone_regex = re.compile(r'\d\d\d\d\d\d\d\d\d\d')
my_phone_number = phone_regex.search(sentence)
print(my_phone_number)

my_phone_number2 = re.search(r'\d\d\d\d\d\d\d\d\d\d', sentence) # this is what is used in the book
print(my_phone_number2)
```

```
year_regex = re.compile(r'\s\d\d\d\d\s')
year = year_regex.search(sentence)

if year:
    print(year.group())
    print(len(year.group()))
```

```
#defining groups
year_regex = re.compile(r'\s(\d\d\d\d)\s')
year = year_regex.search(sentence)

if year:
    print(year.group(1))
    print(len(year.group(1)))
```

```
sentence = 'My name is vimukthini. My mobile number is 0716275565 and my home number is 0312235449. We are living in year 2021 and today is the 25th of November. 0123456789 0770123123 077'

phone_reg = re.compile(r'\d\d\d\d\d\d\d\d\d\d')
mo = phone_reg.search(sentence)

#findall function
mo_all = phone_regex.findall(sentence)

if mo:
  print(mo.group())
  print(mo_all)
```

```
#specify the first 3 digits
phone_reg = re.compile(r'(071|031|077)\d\d\d\d\d\d\d')
mo = phone_reg.search(sentence)
mo_all = phone_reg.findall(sentence)

if mo:
  print(mo.group(1))
  print(mo_all)
```

```
#specify the first 3 digits
phone_reg = re.compile(r'((071|031|077)\d\d\d\d\d\d\d)')
mo = phone_reg.search(sentence)
mo_all = phone_reg.findall(sentence)

if mo:
  print(mo.group())
  print(mo_all)
```

```
s = 'Batman is using his Batmobile as his land vehicle and uses Batcoptor as his plane'

bat_reg = re.compile(r'Batman|Batmobile|Batcoptor')
mo = bat_reg.search(s)
mo_all = bat_reg.findall(s)

if mo:
  print(mo)
  print(mo.group())
  print(mo_all)
```

```
s = 'Batman is using his Batmobile as his land vehicle and uses Batcoptor as his plane'

bat_reg = re.compile(r'(Bat(man|mobile|coptor))')
mo = bat_reg.search(s)
mo_all = bat_reg.findall(s)

if mo:
  print(mo)
  print(mo.group())
  print(mo_all)
```

```
phone_sentence = ' +94716275565 +94312235449'

phone_rg = re.compile(r'((\+94)(\d{9}))')

mo = phone_rg.search(phone_sentence)
mo_all = phone_rg.findall(phone_sentence)

if mo:
  print(mo)
  print(mo.group())
  print(mo.group(1))
  print(mo.group(2))
  print(mo.group(3))
  print(mo_all)
```

```
numbers = 'There are numbers 234 22444 223445 233 2234543'

number_reg =re.compile(r'\s(\d{3})?\s')
mo = number_reg.search(numbers)
mo_all = number_reg.findall(numbers)

if mo:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
number_reg = re.compile(r'(\d{3,6})\s') # {x, y} at least x, or not more than y
mo = number_reg.search(numbers)
mo_all = number_reg.findall(numbers)

if mo:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
numbers = 'There are numbers  1 2 22 234 223445 12345 asdfasdf 233 2234543 sdfasdfsa 12345 '

number_reg = re.compile(r'[0-9]{1,7}\s') 
mo = number_reg.search(numbers)
mo_all = number_reg.findall(numbers)

if mo:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
food_sentence = 'I love to eat at the new food restaurant'
two_vowels = re.compile(r'[aeiou]{2}')
mo = two_vowels.search(food_sentence)
mo_all = two_vowels.findall(food_sentence)

if mo:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
# Character classes
# \d numeric / digits
# \D not digits
# \w alpha numeric
# \W not alpha numeric
# \s space / tab 
# \S not space / tabs

# custom character class
# [0-5a-f] [0-5+-*#]


# ? zero or one occurrence
# * zero or more 
# + one or more

#{x} repeat what's before x times
```

```
s = 'My name is vimukthini my email addresses are vimukthini.jayalath@lseg.com and vimukthinij@gmail.com also I have a hotmail email id which is vimukthini_jayalath@hotmail.com'

email_reg =re.compile(r'[a-z._]+@[a-z._]+')
mo = email_reg.search(s)
mo_all = email_reg.findall(s)

if mo_all:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
my_reg_string = '''
[a-z._]+  #This will capture the username
@         #This will capture the separator 
[a-z._]+  #This will capture the domain
'''
email_reg = re.compile(my_reg_string, flags=re.VERBOSE)
#re.VERBOSE you can add comments and white space to the regex

mo = email_reg.search(s)
mo_all = email_reg.findall(s)

if mo_all:
    print(mo)
    print(mo.group())
    print(mo_all)
```

```
#start and end 

e = ['My Name is vimukthini', 'My home town is Negombo', 'I am 25 years old', 'LSEG BSL is my working place']

my_start_string_reg = re.compile(r'^My') # ^ denotes the start
my_end_string_reg = re.compile(r'vimukthini$') # $ denotes the end

for line in e:

  if my_start_string_reg.search(line):
    print("Found something starting with My")
    print(line)

  if my_end_string_reg.search(line):
    print("Found something ending with vimukthini")
    print(line)

```
